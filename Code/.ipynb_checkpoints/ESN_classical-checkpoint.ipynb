```python
def getPSD(y, N, dt):
    w=blackman(N)
    PSD=(dt/N)*abs(np.fft.rfft(y*w))**2
    f=np.fft.rfftfreq(N,d=dt)
    return f, PSD
```

```python
x, sr = librosa.load(os.path.join(dpath, dfiles[0]))
waveformTime = np.arange(start=0, stop=len(x)/sr, step=1/sr)
figwaveform=plt.figure('view waveform')
librosa.display.waveplot(x)
figwaveform.show()
ipd.Audio(x, rate=sr)
```

```python
%%notify
dt = 1/sr

# Set up static random reservoir
Nhidden = 6000 # Number of hidden variables determine "memory" of reservoir

Nconnections    = 3000 # Number of connections in sparse hidden-to-hidden matrix Whh
spectral_radius = 1.4 # Spectral radius of Whh

#log of hyperparameters
# Nhidden=300, Nconnection=1000
# SR= 0.5: prediction died fast
# SR= 1.1 dies off more slowly, but still a decaying sinusoid
# SR=1.1, Nhidden=10000 (does it need more "bandwidth"? is bandwidth=memory?) This seemed to improve bandwidth and give a little more high frequency response. Low freq still bad
# Try chaning Nconnections instead:
# Nconnection=1100, Nhidden=10000, sr=1.1: now it isnt decaying, but closer to a pure sine. Are Nconnections and SR correlated?
# Nconnection=1100, Nhidden=13000,sr=1.1: by far best one yet. The waveform are still clearly different, but the prediction shares a lot more with the truth. The pred spectrum is too high a 10 Hz, too low at 100 Hz, pretty good at 1000 Hz, and too high at 10000 Hz
# I try to train on longer trace now and predict further. Previously it was Ntrain=1000, Npredict=1000, now I do Ntrain=5000, Npredict=5000, with same hyper parameters
# with 5000 to train and predict, the prediction dies fast again (sounds like a digital drum?)
# I try to do with generated data (a 440 Hz sine wave at the same sampling rate)
# Nconnection=1100, Nhidden=13000,sr=1.1 Ntrain=5000, Npredict=5000 I can predict a 440 Hz sine with a little white noise; normal(loc=0.001,scale=0.05,size=len(t))
# I try a longer trace, Ntrain=Npredict=15000, and the result is frequency drift, after 10-ish periods, the prediction has drifted out of phase with the truth. I try to add some frequency noise to the training data, maybe then it can "catch" this better?
# it didnt make much difference. I'll try to add a second tone, the C# above the 440Hz A I used. I give it the same absolute frequency noise 
# it can predict something with two tones, but it isn't accurate. It still has the problem of getting out of phase. At this point I try to make the Nconnections a bit larger.
#  Nconnection=1500, Nhidden=13000,sr=1.1  Ntrain=Npredict=15000. I also try to increase Ntrans (it had been Ntrans=100) to Ntrans = 300: it seems it hits the actual frequencies a bit better now, but still not very perfect. They sound similar but not identical. The waveforms go in and out of phase, but sometimes are OK
# I try to increase Ntrans further then. I set it to Ntrans =1000. This seems to be the right approach, it gets better. 
# I try to add a third tone, an E to make a A Major chord. Also further increase Ntrans to 2000. This worked more poorly. It didnt fit a three peaks in the spectrum, and produces an increasing amount of harmonics.
# I try to make the network bigger: Nconnection=5000, Nhidden=13000,sr=1.1  Ntrain=Npredict=15000,  Ntrans =5000. This wasnt better than the last. I wanna try smaller network and higher Ntrans.
# Nconnection=1000,Nhidden=2000, sr=1.1, Npredict=500, Ntrans=10000, Ntrain = Ntrans+1000 training was almost instant and as good as previous. Prediction is terrible. Makes a huge jump pretty fast, and doesnt really capture spectrum at all
# Try  Nconnection=1000,Nhidden=5000, sr=1.4, Npredict=500, Ntrans=10000, Ntrain = Ntrans+2000 training perfect(! 1e-16), prediction quite bad but better. Keep moving in thsi direction
# Nconnection=3000,Nhidden=5000,spectral radius=1.5, Npredict=Ntrain/2, Ntrans=10000, Ntrain=Ntrans+2000: it went a bit wild in prediction. The training is still very good. I'll reduce SRa and maybe try more Nhidden rather than Nconnection
# Nconnection = 2000,Nhidden=8000, sr = 1.3, Npredict=Ntrain/2,Ntrans=10000, Ntrain = Ntrans+8000. I also reduced the white noise  sigma to 0.005. Worse training performance, HORRIBLE prediction.
# Nconnection = 3000,Nhidden=6000, sr=1.4,Npredict=Ntrain/2,Ntrans=10000, Ntrain = Ntrans+5000.

Wih = uniform(-1,1,(Nhidden,1)) # Random input-to-hidden matrix
bh  = uniform(-1,1,(Nhidden,1)) # Random bias term
Whh = sparse_esn_reservoir(Nhidden,Nconnections,spectral_radius) # Hidden-to-hidden recurrence
```

```python
%%notify
Ntrans   = 10000   # Number of transient initial steps before training
Ntrain   = Ntrans+5000 # Number of steps to train on
Npredict = int(Ntrain*0.5) # Number of steps for free-running prediction

usemusic=0

if usemusic:
    offset = 18000
    data = np.array(x[offset:-1])
    data = data/max(abs(data))
else:
    f = 440.0
    fthird = 554.365
    ffifth = 659.255
    t=waveformTime
    data = np.sin(2*np.pi*t*(normal(loc=f,scale=0.05,size=len(t))))+ np.sin(2*np.pi*t*(normal(loc=fthird,scale=0.05,size=len(t)))) +np.sin(2*np.pi*t*(normal(loc=ffifth,scale=0.05,size=len(t))))  +normal(loc=0.001,scale=0.005,size=len(t))
    data=data/max(abs(data))

    
# First we split the data into training data (input/labels) and prediction labels
# (To verify that our prediction is correct)
inputs, labels, pred_labels = split_train_label_pred(data,Ntrain,Npredict)

trainingData = labels[Ntrans:Ntrain]
# Second, we initialize the hidden state sequence from the input sequence
H = generate_state_matrix((Whh,Wih,bh),inputs,Ntrans)
# From this, we can do a simple least-squares fit to find the hidden-to-output matrix Woh. Other criteria can be built in to the fit, such as Tikhonov-regularization to avoid over-fitting, or other constraints.
Who = lstsq_stable(H,trainingData)
# Our trained model now simply consists of the static matrices Whh (hidden-hidden), Wih (input-hidden), and bh (bias hidden), 
# which are problem-independent and are unrelated to the data, and the learned hidden-to-output matrix Who.
model = (Whh,Wih,bh,Who)
```

```python

# First, we check: how well did our model predict the past? (How 
# well did we manage to fit the training data)

plt.figure("Training sequence vs learned fit")
plt.plot(waveformTime[Ntrans:Ntrain],Who.dot(H.T),color='r', linewidth=0.6,label='fit')
plt.plot(waveformTime[Ntrans:Ntrain],trainingData,linewidth=0.6,linestyle='dashed',color='g', label='training')
plt.legend()
plt.ylim([-max(abs(trainingData)),max(abs(trainingData))])
plt.show()

plt.figure("Absolute error wrt. training data")
plt.plot(waveformTime[Ntrans:Ntrain],Who.dot(H.T)-trainingData, color='k', label='absolute error')
# plt.ylim([-0.2*max(abs(labels[Ntrans:Ntrain])),0.2*max(abs(labels[Ntrans:Ntrain]))])
plt.legend()
plt.show()


trainFreq, trainPSD = getPSD(Who.dot(H.T),len(Who.dot(H.T)),dt)
truthFreq, truthPSD = getPSD(trainingData,len(trainingData),dt)
# plt.figure('PSD of training and truth')
plt.figure()
# plt.plot(waveformTime[Ntrans:Ntrain],Who.dot(H.T),color='r', linewidth=0.6,label='fit')
# plt.plot(waveformTime[Ntrans:Ntrain],trainingData,linewidth=0.6,linestyle='dashed',color='g', label='training')
plt.plot(trainFreq,trainPSD,label='fit',alpha=0.75,color='r')
plt.plot(truthFreq,truthPSD,label='training',alpha=0.75,color='g')
plt.loglog()
plt.grid()
plt.legend()
plt.show()

ipd.Audio(Who.dot(H.T), rate=sr)
# ipd.Audio(trainingData, rate=sr)
```

```python
%%notify
# prediction
# All right, looks pretty good! Maximum error of about 0.00001 in the training set.
# We finally validate by checking if we can predict 500 steps into the future:
state0 = H[-1]
input0 = labels[-1]
# Predict Npredict steps into the future, running free from final
# training input and hidden state.
ys, h = predict(model,state0,input0,int(Npredict))
```

```python
#plot prediction
predTime=waveformTime[Ntrain:Ntrain+int(Npredict)]
# Let's compare the 500-step forecast with the true continuation of the sequence
plt.figure()
plt.plot(predTime,pred_labels,color='g', alpha=0.5,linewidth=0.5,label='truth')
plt.plot(predTime,ys,color='r', alpha=0.5, linewidth=0.5,label='prediction')
plt.legend()
# plt.ylim([-1,1])
plt.ylim([-max(abs(pred_labels)),max(abs(pred_labels))])
# plt.title("500-step prediction vs truth")
plt.show()

plt.figure()
plt.plot(predTime,(ys-pred_labels),'k',linewidth=0.5,alpha=0.5)
plt.title("Absolute error prediction vs truth")
plt.show()

plt.figure()
predFreq, predPSD = getPSD(ys,len(ys),dt)
truthFreq, truthPSD = getPSD(pred_labels,len(ys),dt)
plt.plot(predFreq,predPSD,label='prediction',alpha=0.5,color='r')
plt.plot(truthFreq,truthPSD,label='truth',alpha=0.5,color='g')
plt.loglog()
plt.grid()
plt.legend()
```
